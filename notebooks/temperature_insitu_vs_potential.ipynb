```python
import warnings
import iris

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    cubes = iris.load('woa01an.nc')

print(cubes)
```

```python
from iris import Constraint

x, y = 142.2, 11.35

dx = dy = 0.5
lat = Constraint(latitude=lambda cell: y-dy <= cell <= y+dy)
lon = Constraint(longitude=lambda cell: x-dx <= cell <= x+dx)

temp = cubes[8].extract(lon & lat)
salt = cubes[5].extract(lon & lat)
```

```python
import gsw


t, p = temp.data, temp.coord('Depth').points

SA = gsw.SA_from_SP(salt.data, p, lon=x, lat=y)

CT = gsw.CT_from_t(SA, t, p)
```

```python
%matplotlib inline
from matplotlib import style
import matplotlib.pyplot as plt


style.use('dark_background')

fig, ax = plt.subplots(figsize=(6, 8))
ax.invert_yaxis()
ax.plot(t, p, linewidth=2)
ax.plot(CT, p, linewidth=2)
ax.set_xlabel(u'Temperatura [\u00B0C]')
ax.set_ylabel('Profundidade [m]')

# Zoom.
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes

axins = zoomed_inset_axes(ax, 2, loc=5)
axins.plot(t, p)
axins.plot(CT, p, linewidth=2)
axins.invert_yaxis()
axins.axis([0.25, 3.25, 5500, 3000])
axins.xaxis.tick_top()
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")
axins.xaxis.set_major_locator(MaxNLocator(nbins=1, prune='lower'))
axins.yaxis.set_major_locator(MaxNLocator(nbins=4, prune='lower'))
```
